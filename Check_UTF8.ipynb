```python
# -*- coding: utf-8 -*-
#
# SCRIPT DE DIAGNÃ“STICO DE ENCODING: check_encoding.py
#

import pandas as pd

# --- CONFIGURE AQUI ---
# Coloque o caminho exato para o seu arquivo CSV
CSV_PATH = r"C:\REPOSITORIO\nome_do_arquivo.csv"
# --------------------

# Lista de codificaÃ§Ãµes mais comuns para testar
ENCODINGS_TO_TRY = [
    'utf-8',
    'latin-1',
    'iso-8859-1', # Nome formal do latin-1
    'cp1252',     # Windows Latin 1
]

print(f"ğŸ•µï¸  Analisando o arquivo: {CSV_PATH}\n")

# Tenta ler as 10 primeiras linhas do arquivo com cada encoding
for encoding in ENCODINGS_TO_TRY:
    print("-" * 50)
    print(f"Tentando ler com a codificaÃ§Ã£o: '{encoding}'")
    
    try:
        # Usamos o leitor de arquivos do Python que Ã© mais sensÃ­vel a erros
        with open(CSV_PATH, 'r', encoding=encoding) as f:
            for i, line in enumerate(f):
                if i >= 10: # Limita a 10 linhas
                    break
                print(f"  Linha {i+1}: {line.strip()}")
        print(f"\nâœ… SUCESSO: O arquivo foi lido com '{encoding}'. Verifique visualmente se a acentuaÃ§Ã£o estÃ¡ correta.")

    except UnicodeDecodeError as e:
        print(f"\nâŒ FALHA: A codificaÃ§Ã£o '{encoding}' nÃ£o Ã© a correta.")
        print(f"   Erro: {e}")
        
    except Exception as e:
        print(f"\nâŒ ERRO INESPERADO com '{encoding}': {e}")

print("-" * 50)
print("\nğŸ” AnÃ¡lise concluÃ­da. Verifique os resultados acima.")
```
